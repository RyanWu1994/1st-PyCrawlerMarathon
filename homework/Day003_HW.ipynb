### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

fh = open("./Data/64_72hr_CH.xml", "r",encoding='utf-8')
xml = fh.read()
fh.close()

d = xmltodict.parse(xml)
print(len(d['cwbopendata']['dataset']['locations']['location']),'個地區')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for i in d['cwbopendata']['dataset']['locations']['location']:
    print('地區',i['locationName'])
    print('時間',i['weatherElement'][0]['time'][0]['dataTime'])
    print('溫度',i['weatherElement'][0]['time'][0]['elementValue']['value'])
    print('====================================================')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

for i in d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time']:
    print(i['dataTime'],f"攝氏度{i['elementValue']['value']}度")
    
    
```
